```python
import time
import matplotlib.pyplot as plt
```

```python
k_p = 0.1
k_i = 0.0018
k_d = 0.1
interval = 1

setpoint = 100 # 100 km destination
values = [0, 10, 20, 30, 50, 70, 100] # how far reached
error = setpoint # initially

i = 0
y = []

error_prev = 0 
integral = 0
while (error != 0): 
    
    val = values[i]
    i += 1

    # calculate PID terms
    error = setpoint -val
    integral += error * interval 
    derivative = (error - error_prev) / interval
    output = (k_p * error) + (k_i * integral) + (k_d * derivative)

    error_prev = error
    y.append(output)

    # wait for interval time
    time.sleep(interval)
    print(f'error = {error}, required_pedal = {output} ')
```

```python
t = [0, 1, 2, 3, 4, 5, 6]
plt.plot(t, y)
```
