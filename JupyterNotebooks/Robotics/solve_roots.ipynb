```python
from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return x**2-x-2

root = fsolve(func, 0)
print(root)
```

```python
x = np.arange(-6, 6, 0.1)
y = func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.title('Plot of the function')
plt.grid(color='g', linestyle='--')
```
