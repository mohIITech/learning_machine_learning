### Write a python code to read the data points from the .txt file and find the covariance matrix
C ∈ R
d×d
for this data.

```python
# reading the data
df = pd.read_csv('/content/drive/MyDrive/Classroom/CSL2010: Introduction to Machine Learning Institute Core/iris.txt', sep=',')

# get a view
df.head()
```

```python
# make the covariance matrix
cov_mat = df[['sepal length', 'sepal width', 'petal length', 'petal width']].cov()
cov_mat
```

### Find the pairs of features which are most (positively/negatively) correlated and the least correlated.

```python
# copying cov_mat with no diagonal values
cov_mat_no_diag = cov_mat.copy()
np.fill_diagonal(cov_mat_no_diag.values, np.nan)

# finding most positive and negative covariance
max_cov = cov_mat_no_diag.max().max()
min_cov = cov_mat_no_diag.min().min()

# least correlated
abs_cov = cov_mat_no_diag.abs() # applies mod
min_abs_cov = abs_cov.min().min()

# finding the pairs of features
max_pair = np.where(cov_mat_no_diag == max_cov)
min_pair = np.where(cov_mat_no_diag == min_cov)
min_abs_pair = np.where(abs_cov == min_abs_cov)
```

```python
max_pair_name = df.columns[max_pair[0][1]], df.columns[max_pair[1][1]]
min_pair_name = df.columns[min_pair[0][1]], df.columns[min_pair[1][1]]
min_abs_pair_name = df.columns[min_abs_pair[0][1]], df.columns[min_abs_pair[1][1]]
```

### This gives all the pairs

```python
print(f'max_pair_name = {max_pair_name}')
print(f'min_pair_name = {min_pair_name}')
print(f'min_abs_pair_name = {min_abs_pair_name}')
```

### Find the eigenvalues of the covariance matrix. Observe the relationship between the variance of the i-th feature and the i-th eigenvalue of the covariance matrix.

```python
eigenvalues, eigenvectors = np.linalg.eigh(cov_mat)
eigenvalues
```

```python
sum = 0
for i in range(len(eigenvalues)):
  sum += eigenvalues[i]

sum
```
